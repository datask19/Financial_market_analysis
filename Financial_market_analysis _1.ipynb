```python
# Downloading stock data for Microsoft (MSFT) from Yahoo Finance.
start_date = '2019-9-9'
end_date = '2023-5-5'
symbol = 'MSFT'
data = yf.download(symbol, start=start_date, end=end_date)
print(data.head())
```

```python
# Calculating daily returns and plotting a histogram.
returns = data['Adj Close'].pct_change()
plt.hist(returns.dropna(), bins=30 ,edgecolor='black')
plt.title('Histogram of Daily Returns for MSFT')
plt.xlabel('Daily Returns')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the closing prices over the given time period.
data['Adj Close'].plot(figsize=(12, 6),color ='r')
plt.title('MSFT Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.show()

# Checking for trends or patterns in the stock prices using a rolling mean.
data['Adj Close'].plot(label='Adj Close', figsize=(12, 6))
data['Adj Close'].rolling(window=30).mean().plot(label='30-Day Avg', color='g')
plt.title('MSFT Stock Prices with 30-Day Rolling Average')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.legend()
plt.show()
```
